```python
%matplotlib inline
import numpy as np 
import matplotlib.pyplot as plt 
import math


shift=.08
shift0=0

width=792
height=600
Nx=np.arange(1, width)
Ny=np.arange(1, height)
xcoord=Nx - width/2
ycoord=-(Ny - height/2)
[Xc,Yc]=np.meshgrid(xcoord,ycoord)

Modetot=1

shift0=0
phase=np.remainder(np.angle(Modetot)-shift*(np.sqrt(Xc**2+Yc**2))-shift0*(Xc),2*(math.pi))/2*(math.pi)*255

```

```python
plt.imshow(phase)

```
